```python
url = input("Enter URL: ").lower()

score = 0

if "@" in url:
    score += 1

if "-" in url:
    score += 1

if url.count(".") > 3:
    score += 1

if url.startswith("http://"):
    score += 1

keywords = ["login", "verify", "update", "bank", "secure", "account"]

for word in keywords:
    if word in url:
        score += 1

if score >= 2:
    print("Warning: This URL is likely a phishing link.")
else:
    print("This URL appears to be safe.")
```
