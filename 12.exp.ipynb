```python
from datetime import datetime

security_events = [
    {
        "Time": "2026-07-21 09:10:15",
        "Event": "Failed Login",
        "Severity": "Medium",
        "Source IP": "192.168.1.15"
    },
    {
        "Time": "2026-07-21 09:20:40",
        "Event": "Port Scan Detected",
        "Severity": "High",
        "Source IP": "10.0.0.8"
    },
    {
        "Time": "2026-07-21 09:35:12",
        "Event": "Malware Signature Detected",
        "Severity": "Critical",
        "Source IP": "172.16.0.20"
    }
]

# Generate Incident Report
print("=" * 60)
print("        CYBER SECURITY INCIDENT REPORT")
print("=" * 60)
print("Report Generated:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print()

for i, event in enumerate(security_events, start=1):
    print(f"Incident {i}")
    print("-" * 40)
    print("Time       :", event["Time"])
    print("Event      :", event["Event"])
    print("Severity   :", event["Severity"])
    print("Source IP  :", event["Source IP"])
    print()

print("=" * 60)
print("Summary")
print("=" * 60)
print("Total Incidents :", len(security_events))

critical = sum(1 for e in security_events if e["Severity"] == "Critical")
high = sum(1 for e in security_events if e["Severity"] == "High")
medium = sum(1 for e in security_events if e["Severity"] == "Medium")
low = sum(1 for e in security_events if e["Severity"] == "Low")

print("Critical :", critical)
print("High     :", high)
print("Medium   :", medium)
print("Low      :", low)
print("=" * 60)
```
