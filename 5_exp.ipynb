```python
import socket

domains = input("Enter domain names separated by commas: ").split(",")

print("\nDomain Name\t\tIP Address")
print("-" * 40)

for domain in domains:
    domain = domain.strip()
    try:
        ip = socket.gethostbyname(domain)
        print(domain, "\t", ip)
    except:
        print(domain, "\t Unable to resolve")
```
