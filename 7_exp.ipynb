```python
log_data = [
    "192.168.1.10 FAILED",
    "192.168.1.20 SUCCESS",
    "192.168.1.10 FAILED",
    "192.168.1.30 FAILED",
    "192.168.1.20 FAILED",
    "192.168.1.10 SUCCESS"
]

failed_count = {}

for entry in log_data:
    ip, status = entry.split()

    if status == "FAILED":
        if ip in failed_count:
            failed_count[ip] += 1
        else:
            failed_count[ip] = 1

print("Failed Login Attempts per IP:\n")

for ip, count in failed_count.items():
    print(ip, ":", count)
```
