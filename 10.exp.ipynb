```python

login_logs = [
    ("192.168.1.10", "FAILED"),
    ("192.168.1.20", "SUCCESS"),
    ("192.168.1.10", "FAILED"),
    ("192.168.1.30", "FAILED"),
    ("192.168.1.10", "FAILED"),
    ("192.168.1.30", "FAILED"),
    ("192.168.1.30", "FAILED"),
]

threshold = 3

failed_attempts = {}

print("Analyzing Login Logs...\n")

for ip, status in login_logs:
    if status == "FAILED":
        failed_attempts[ip] = failed_attempts.get(ip, 0) + 1

for ip, count in failed_attempts.items():
    print(f"{ip} -> Failed Attempts: {count}")

    if count >= threshold:
        print(f" ALERT: Possible Intrusion Detected from {ip}!\n")
```
