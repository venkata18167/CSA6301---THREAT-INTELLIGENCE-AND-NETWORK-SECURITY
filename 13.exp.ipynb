```python
# ==========================================
# File Integrity Checker using SHA-256
# Google Colab Compatible
# ==========================================

import hashlib
from google.colab import files

# Function to calculate SHA-256 hash
def calculate_hash(file_path):
    sha256 = hashlib.sha256()

    with open(file_path, "rb") as f:
        while True:
            data = f.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

# -------------------------------
# Upload Original File
# -------------------------------
print("Upload the ORIGINAL file")
uploaded = files.upload()

original_file = list(uploaded.keys())[0]

baseline_hash = calculate_hash(original_file)

print("\nOriginal File:", original_file)
print("Baseline SHA-256 Hash:")
print(baseline_hash)

# -------------------------------
# Upload File Again
# -------------------------------
input("\nPress Enter after modifying the file (if desired)...")

print("\nUpload the SAME file again (modified or unmodified)")
uploaded2 = files.upload()

new_file = list(uploaded2.keys())[0]

current_hash = calculate_hash(new_file)

print("\nCurrent SHA-256 Hash:")
print(current_hash)

# -------------------------------
# Compare Hashes
# -------------------------------
print("\n========== RESULT ==========")

if baseline_hash == current_hash:
    print("File Integrity Verified")
    print("The file has NOT been modified.")
else:
    print("WARNING!")
    print("The file HAS been modified.")

print("============================")
```
