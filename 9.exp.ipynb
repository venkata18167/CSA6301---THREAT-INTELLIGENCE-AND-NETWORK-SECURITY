```python

blocked_ips = ["192.168.1.100", "10.0.0.5"]
blocked_ports = [23, 25]  # Telnet and SMTP
allowed_protocols = ["TCP", "UDP"]

packet = {
    "source_ip": "192.168.1.100",
    "destination_ip": "192.168.1.10",
    "port": 80,
    "protocol": "TCP"
}
print("Checking Packet...\n")

if packet["source_ip"] in blocked_ips:
    print("Result: BLOCKED")
    print("Reason: Source IP is blocked.")

elif packet["port"] in blocked_ports:
    print("Result: BLOCKED")
    print("Reason: Port is blocked.")

elif packet["protocol"] not in allowed_protocols:
    print("Result: BLOCKED")
    print("Reason: Protocol is not allowed.")

else:
    print("Result: ALLOWED")
    print("Packet passed the firewall.")

print("\nPacket Details:")
for key, value in packet.items():
    print(f"{key}: {value}")
```
