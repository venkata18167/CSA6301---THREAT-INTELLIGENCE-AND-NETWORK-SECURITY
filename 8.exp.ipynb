```python
import hashlib

target_hash = hashlib.sha256("password123".encode()).hexdigest()

wordlist = [
    "admin",
    "welcome",
    "letmein",
    "123456",
    "password",
    "password123",
    "qwerty",
    "hello"
]

print("Target Hash:")
print(target_hash)
print("\nStarting Dictionary Attack...\n")

found = False

for word in wordlist:
    hashed_word = hashlib.sha256(word.encode()).hexdigest()

    print(f"Trying: {word}")

    if hashed_word == target_hash:
        print("\nPassword Found!")
        print("Password:", word)
        found = True
        break

if not found:
    print("\nPassword not found in the wordlist.")
```
