```python
malicious_ips = [
    "192.168.1.10",
    "10.0.0.5",
    "203.0.113.25"
]

traffic_log = [
    "192.168.1.2",
    "192.168.1.10",
    "8.8.8.8",
    "10.0.0.5",
    "172.16.0.1"
]

print("Scanning Traffic Log...\n")

for ip in traffic_log:
    if ip in malicious_ips:
        print(ip, "-> Malicious IP Detected")
    else:
        print(ip, "-> Safe")
```
